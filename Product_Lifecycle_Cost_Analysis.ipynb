```python
import pandas as pd

# Load the CSV file
data = pd.read_csv('product_data_large.csv')
# Display the first few rows
data.head()
```

```python
import matplotlib.pyplot as plt

# Filter data for a specific product, e.g., "Mobile"
mobile_data = data[data['Product'] == 'Mobile']

# Create a bar chart
plt.bar(mobile_data['Stage'], mobile_data['Cost_in_Stage'], color='blue')
plt.title('Cost in Each Lifecycle Stage for Mobile')
plt.xlabel('Stage')
plt.ylabel('Cost')
plt.show()
```
